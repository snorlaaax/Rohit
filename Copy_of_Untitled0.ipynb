```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Dummy Sales Dataset
data = {
    "Date": ["2025-01-02", "2025-01-05", "2025-01-15", "2025-02-10", "2025-02-12", "2025-02-20", "2025-03-05", "2025-03-15"],
    "Product": ["Laptop", "Phone", "Headphone", "Laptop", "Phone", "Headphone", "Laptop", "Phone"],
    "Quantity": [2, 5, 10, 1, 3, 8, 1, 4],
    "Price": [50000, 15000, 2000, 52000, 16000, 1800, 55000, 15500]
}

df = pd.DataFrame(data)

# Step 2: Total Sales Column
df['Total_Sales'] = df['Quantity'] * df['Price']

# Step 3: Summary
print("Total Revenue:", df['Total_Sales'].sum())
print("Average Order Value:", df['Total_Sales'].mean())

# Step 4: Product-wise Sales
product_sales = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
print("\nProduct-wise Sales:\n", product_sales)

# Step 5: Monthly Sales
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total_Sales'].sum()
print("\nMonthly Sales:\n", monthly_sales)

# Step 6: Graphs
# Product-wise Sales
product_sales.plot(kind='bar', title="Product-wise Sales", figsize=(7,5))
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.show()

# Monthly Sales Trend
monthly_sales.plot(kind='line', marker='o', title="Monthly Sales Trend", figsize=(7,5))
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```
